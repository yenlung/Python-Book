```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

接著下來我們來生成一堆的數字。我們從「平均值是 0, 標準差是 1」的標準常態分佈中, 抽出 $n$ 個樣本。這神奇指令是:
    
    np.random.randn(n)

```python
np.random.randn(10)
```

我們抽 100 個樣本畫出來!

```python
plt.plot(np.random.randn(100))
```
