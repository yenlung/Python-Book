現在我們來畫個圖!

```python
π = np.pi
```

```python
θ = np.linspace(0, 2*π, 500)
```

```python
r = 3
x = r * np.cos(θ)
y = r * np.sin(θ)
```

```python
plt.plot(x, y)
```

這根本不圓啊!! 我們要來調整一下。

這時, 需要知道 `matplotlib` 畫圖的基本架構。

![matplotlib 圖的結構](https://github.com/yenlung/Python-Book/blob/master/images/figure_axes.png?raw=1)

我們要設定一組 axes 的屬性, 就需要知道目前在用的 axes 是哪一組。這要怎麼知道呢? 就是用

    ax = plt.gca()
    
`gca` 的意思是 "Get Current Axes"。

```python
ax = plt.gca()
ax.set_aspect('equal')

plt.plot(x, y)
```

果然成功了!

附帶一提, 我們之所以用了 `r=3` 這樣的設定, 是因為如此很容易造出「可變半徑」的極座標系統。比方說, 我們想畫出:

$$r = 1 - \sin(\theta),$$

可以這樣畫出來...

```python
r = 1 - np.sin(θ)

x = r * np.cos(θ)
y = r * np.sin(θ)
```

```python
ax = plt.gca()
ax.set_aspect('equal')

plt.plot(x, y)
```

但「藍心」看來怪怪的, `matplotlib` 有可「快速上色法」, 用一個字母就代表一個顏色。

![matplotlib 快速顏色](https://github.com/yenlung/Python-Book/blob/master/images/matplotlib_quick_color.png?raw=1)

```python
ax = plt.gca()
ax.set_aspect('equal')

plt.plot(x, y, 'r')
```
