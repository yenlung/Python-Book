```
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 1. 善用 enumerate

```
L = ['a', 'b', 'c']

for i in L:
    print(i)
```

加上編號。

```
for i in range(3):
    print(i+1, L[i])
```

試試 `enumerate` 會做什麼。

```
list(enumerate(L))
```

用 `for` 迴圈一一顯示出來。

```
for i in enumerate(L):
    print(i)
```

用 unpacking 取出內容, 再修正編號從 1 開始。

```
for i, s in enumerate(L):
    print(i+1, s)
```

### 2. 畫多個圖

```
x = np.linspace(-10, 10, 200)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

用 $2 \times 2$ 的排列方式畫圖。

```
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))

plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))

plt.subplot(2, 2, 3)
plt.plot(x, x)

plt.subplot(2, 2, 4)
plt.plot(x, x**2)
```

把加州房價的資料拿進來

```
from sklearn.datasets import fetch_california_housing
cal = fetch_california_housing()
```

回顧一下所有 features 的名稱。

```
cal.feature_names
```

看各個 feature 和房價的闗係圖。

```
X = cal.data
Y = cal.target

plt.figure(figsize=(8, 10))
for i, feature in enumerate(cal.feature_names):
    plt.subplot(5, 3, i+1)
    plt.scatter(X[:, i], Y, s=1)
    plt.ylabel('price')
    plt.xlabel(feature)
    plt.tight_layout()
```
