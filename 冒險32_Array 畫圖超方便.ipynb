用 array 畫函數。

先設 x 座標。

```python
x = np.linspace(0, 20, 100)
```

指定畫哪種函數。

```python
plt.plot(x, np.sin(x))
```

同時畫兩種函數也可以，還會自動用不同顏色標示。

```python
plt.plot(x, np.sin(x)+x)
plt.plot(x, np.cos(x)-x)
```
