```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

接著我們開始畫函數! 畫函數基本上就是「描點法」!

![np.linspace](https://github.com/yenlung/Python-Book/blob/master/images/linspace.png?raw=1)

假設我們要畫的第一個函數是

$$f(x) = x^2 + 1$$

X 軸的範圍取 [-10, 10], 取 500 個點。

```python
x = np.linspace(-10, 10, 500)
```

```python
y = x**2 + 1
```

```python
plt.plot(x, y)
```

再來我們畫

$$f(x) = \sin(x)$$

X 軸的範圍取 [0, 6π], 1000 個點。

```python
x = np.linspace(0, 6*np.pi, 1000)
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
```
